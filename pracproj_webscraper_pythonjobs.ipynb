# Web scraping https://pythonjobs.github.io

Doing absolutely same shit just different website, to better memorise the process.


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


URL = "https://pythonjobs.github.io/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

soup
```

# Header

```python
header = soup.find("header")
header.find("div", id="title").text.strip() + " " + header.find("div", class_="sub").text.strip()
```

# Main content of the site

```python
results = soup.find("div", id="main").find("section", class_="job_list")
job_elements = results.find_all("div", class_="job")

job_titles: list = []
job_locations: list = []
job_publish_date: list = []
all_urls: list = []
job_companies: list = []
for job in job_elements:
  job_titles.append(job.find("h1").find("a").text.strip())

  job_locations.append(job.find("i", class_="i-globe").parent.text.strip())

  job_publish_date.append(job.find("i", class_="i-calendar").parent.text.strip())

  job_companies.append(job.find("i", class_="i-company").parent.text.strip())

  all_urls.append(job.find("a")["href"])

dictpd = {
    "Title": job_titles,
    "Company": job_companies,
    "Location": job_locations,
    "Publish Date": job_publish_date,
    "URL": [URL+url for url in all_urls]
}

df = pd.DataFrame(dictpd)
df
```
